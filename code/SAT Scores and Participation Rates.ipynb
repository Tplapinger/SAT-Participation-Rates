# SAT Scores and Participation Rates

Load and print data

```python
import csv           ##Format copied and adapted from 1.05-iteration-csvs lesson
data=[]                ###initializing data list
with open("data/sat_scores.csv", "rU") as f: ## importing csv file
    reader = csv.reader(f)
    for row in reader:
        data.append(row)
```

```python
for i in data:      
    print i             #printing the data
```

Split List By Column Names and Data

```python
col_names=data[0]   ##Names of the Columns(ie. State, Rate, Verbal, Math)
dataset=data[1:-1]    ###List of data seperate from column headers and seperating the ALL column so later calculations are correct

print dataset
```

Divide State, Particpation Rate, Verbal and Mathmatics Scores into Seperate Lists

```python
state_names=[column[0] for column in dataset] #Creating a list of States from dataset
print state_names
```

```python
rate=[column[1] for column in dataset] #Creating a list of Participation Rates from dataset in order to ease later calculation
print rate
```

```python
verbal=[column[2] for column in dataset] #Creating a list of Verbal Scores from dataset in order to ease later calculation
print verbal
```

```python
maths=[column[3] for column in dataset]
print maths
```

Inspect and confirm The Type of Data

```python
print type(state_names) #State Column is a List of strings
print type(state_names[0])
print type(rate)    #Rate Column is a List of strings
print type(rate[0])
print type(verbal)  #Verbal Column is a List of strings
print type(verbal[0])
print type(maths)   #Math Column is a List of strings
print type(maths[0])
```

Converting Rate, Verbal, and Maths to Integers

```python
#State is fine as a list of strings, but in order to calculate summary statistics 
#later Rate, Verbal, and Maths have to be converted to list of integers. 

rate_int = [int(x) for x in rate] #conversion of rate to list of integers
print rate_int
verbal_int = [int(x) for x in verbal] #conversion of verbal scores to list of integers
print verbal_int
maths_int = [int(x) for x in maths] #conversion of maths scores to list of integers
print maths_int
```

Creating dictionary for each column mapping the State to its respective value

```python
state_rate_dict= dict(zip(state_names,rate_int)) #creation of dictionary pairing key, State, to value, participation rate. 
print state_rate_dict
state_verbal_dict= dict(zip(state_names,verbal_int)) #creation of dictionary pairing key, State, to value, verbal scores. 
print state_verbal_dict
state_maths_dict=dict(zip(state_names,maths_int)) #creation of dictionary pairing key, State, to value, maths scores. 
print state_maths_dict
```

Dictionary with the values for each of the numeric columns

```python
numerics= [rate_int, verbal_int, maths_int] #creating singular list of all three numeric columns
numeric_dict=dict(zip(col_names[1:5], numerics)) #creating dictionary connecting keys Rate, Math, and Verbal to values in their respecive columns 
print numeric_dict
```

### Summary Statistics

```python
#Printing out minimum and maximum of each numeric column
print "Rate Minimum is " + str(min(rate_int)) 
print "Rate Maximum is " + str(max(rate_int))
print "Verbal Minimum is " + str(min(verbal_int))
print "Verbal Maximum is " + str(max(verbal_int))
print "Maths Minimum is " + str(min(maths_int))
print "Maths Maximum is " + str(max(maths_int))
```

```python
def stddev(data):             #defining function stddev()
    mean=sum(data)/len(data)     #computing mean = Sum of the the inputed data divided by its count
    square=[(x-mean)**2 for x in data]   #calculating the deviations of each data point in data and Squaring it
    var=sum(square)/len(square)      #computing variance by finding average of square
    std_dev= var ** 0.5            #square rooting the variance to become standard deviation
    return std_dev
```

```python
#Applying the stddev function to print Numeric Columns indiviudal Stadard deviation
print "Rate Standard Deviation is " + str(stddev(rate_int))
print "Verbal Standard Deviation is " +str(stddev(verbal_int))
print "Maths Standard Deviation is " + str(stddev(maths_int))
```

```python
#Numpy confirmation of correct standard deviation, means, and variance
import numpy as np
print np.mean(rate_int)
print np.mean(verbal_int)
print np.mean(maths_int)
print np.std(rate_int)
print np.std(verbal_int)
print np.std(maths_int)
```
